A simple driver notebook that shows how to access training and validation data

```python
import os
import torch

from NeuralExec.utility import read_pickle
from NeuralExec.llm import load_llm
from NeuralExec.ex_triggers import NeuralExec

# load data
training_prompts = read_pickle('./data/prompts_training.pickle')
# load model
llm = load_llm('mistralai/Mistral-7B-Instruct-v0.2')
# load an example trigger 
logger = read_pickle('./logs/15_5_mistral')
ne, _ = logger.get_last_adv_tok(best=True)
```

Example of training prompt without target string:

```python
prompt = training_prompts[0]

prompt_str = prompt(llm, ne, with_target=False)
print(prompt_str)
```

Example of training prompt with target string:

```python
prompt_str = prompt(llm, ne, with_target=True)
print(prompt_str)
```

# Printing some more

```python
n = 10
for i in range(n):
    prompt = training_prompts[i+1]
    prompt_str = prompt(llm, ne, with_target=False)
    print(prompt_str, end='\n\n'+'_#_'*30+'\n\n')
```
